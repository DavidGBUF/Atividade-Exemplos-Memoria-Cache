    Cálculo da quantidade de bits necessários para uma determinada memória cache.
Considere um sistema de computação  com uma memória cache de 32KB de capacidade, constituída de linhas de 8 bytes de largura. A MP possui uma capacidade de 16MB.

Resolução: 
    A Quantidade total de total de bits que a memória cache pode trabalhar vai ser determinada
por:
    O n° de bits de dados que ela armazena + o n° de bits necessários para representar todos os
blocos de dados que ela armazena.

*O Primeiro pode ser facilmente obtido por:
Se a memória tem capacidade de 32Kbytes e cada byte equivale a 8 bits então basta multiplicar:
32Kx8= 256K bits de dados

    *O segundo pode ser obtido por:
Quantos bits de tag uma linha possui multiplicado pelo número de linhas.
O número de linhas é dado por: Capacidade da cache/ n° de células por linha.
2^15/2³= 2¹² linhas, ou seja 12 bits para representar uma linha. 


O tamanho do endereço que a cache trabalha vai ser o mesmo tamanho do endereço da MP
A MP possui 16MB, ou seja 16M células (cada célula tem 1 byte) , ou seja 16M endereços
16M=2^24 endereços, para representar esses endereços é necessário 24 bits 

Desses 24 bits: 12 sinalizam a linha da cache, 3 representam a célula do bloco armazenado (8 células por bloco, buscar dado no enunciado), então para completar 24 serão necessários 9 bits de tag, se são 9 bits por tag e cada linha possui uma tag, então bits por tag x numero de linhas= numero de bits das tags:

9 X 2¹² = 36K bits de tags

Somando os bits totais:
36K bits de tag + 256K bits de dados= 292K bits 

A memória cache vai ter que trabalahar com 292K bits
